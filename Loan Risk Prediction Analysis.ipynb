# Data visuals

### Loan status bar chart

This visualization reveals that the dataset is imbalanced, with approximately 68% of loans being approved (Y) and 32% rejected (N).

```python
# grid style
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))

# Create the count plot
ax = sns.countplot(x='Loan_Status', data=df, palette='viridis', order=df['Loan_Status'].value_counts().index)

# Add title and labels
plt.title('Distribution of Loan Approval Status', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('Loan Approval Status (Y: Yes, N: No)', fontsize=14)
plt.ylabel('Number of Applicants', fontsize=14)

# Add counts on top of the bars 
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points',
                   fontsize=12, fontweight='bold')

# Show the plot
plt.tight_layout()
plt.show()
```

### Correlation heatmap

The heatmap identifies the linear relationships between variables. Most notably, we see a moderate positive correlation between LoanAmount and ApplicantIncome.

# Feature Engineering

### Encoding

```python
df1=pd.get_dummies(df,columns=["Gender","Married","Dependents","Education","Self_Employed","Property_Area","Loan_Status"])
```

```python
df1
```

```python
df1.dtypes
```

```python
df1.drop("Gender_Female",axis=1,inplace=True)
df1.drop("Married_No",axis=1,inplace=True)
df1.drop("Education_Graduate",axis=1,inplace=True)
df1.drop("Self_Employed_No",axis=1,inplace=True)
df1.drop("Property_Area_Rural",axis=1,inplace=True)
df1.drop("Loan_Status_N",axis=1,inplace=True)
df1.drop("Dependents_3+",axis=1,inplace=True)
```

```python
df1
```

```python
x=df1.iloc[:,2:15]
x
```

```python
y=df1.iloc[:,-1]
y
```

# Logistic Regression

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=(0))
```

```python
x_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lm=LogisticRegression()
lm.fit(x_train,y_train)
```

```python
y_pred=lm.predict(x_test)
y_pred
```

```python
y_test
```

### Confusion matrix

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
#131 correct prediction and 23 wrong prediction
```

```python
from sklearn.metrics import classification_report
```

```python
# Generating the detailed report
print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=['Loan Denied (N)', 'Loan Approved (Y)']))
```

### Summary

Our model achieved an accuracy of 85.06%. Most importantly, it only produced 3 False Positives, demonstrating that the system is highly effective at minimizing financial risk by avoiding bad loans. Optimized model for business growth, achieving a 97% recall rate for approved loans, ensuring minimal loss of eligible customers. Mitigated financial risk by maintaining a high precision rate (88%) for loan denials, effectively filtering high-risk applicants.

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Initialize the model
rf = RandomForestClassifier(n_estimators=100, random_state=0)
```

```python
# Fit the model
rf.fit(x_train, y_train)
# Make predictions
rf_pred = rf.predict(x_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score

print(f"Random Forest Accuracy: {accuracy_score(y_test, rf_pred):.4f}")
print("-" * 30)
print(classification_report(y_test, rf_pred))
```

### Summary

Accuracy: $81.82\%$Performance on Approved Loans: High Recall ($0.92$) and Precision ($0.84$), meaning the model remains very effective at identifying eligible customers.Performance on Denied Loans: The recall for Denied loans improved slightly to $0.56$ 

# Conclusion

While the Logistic Regression model provided higher overall accuracy and better precision, the Random Forest showed a slight improvement in detecting risky (Denied) applications. However, for this specific dataset, the simpler Logistic Regression model generalizes better and provides a more reliable balance of metrics.
